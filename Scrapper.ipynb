```python
from selenium import webdriver
import pandas as pd 
import time
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome(executable_path= 'C:\webdrivers\chromedriver.exe')
driver.maximize_window()
driver.get("https://perizinan.kkp.go.id/portal/grid.php?target=aktif&doc=6&q=")
```

```python
# Perorangan / badan hukum
badan_hukum = driver.find_elements_by_xpath('//tbody/tr/td[1]')
# Nomor Siup
nomor_siup = driver.find_elements_by_xpath('//tbody/tr/td[2]')
# Nama Kapal
nama_kapal = driver.find_elements_by_xpath('//tbody/tr/td[3]')
# Nomor SIPI/SIKPI
nmr_sipi_sikpi = driver.find_elements_by_xpath('//tbody/tr/td[4]')
# Tanggal Terbit
tanggal_terbit = driver.find_elements_by_xpath('//tbody/tr/td[5]')
# Tanggal Berlaku
tanggal_berlaku = driver.find_elements_by_xpath('//tbody/tr/td[6]')
# Alat Tangkap
alat_tangkap = driver.find_elements_by_xpath('//tbody/tr/td[7]')
# GT Kapal
gt_kapal = driver.find_elements_by_xpath('//tbody/tr/td[8]')
# Tanda Selar
tanda_selar = driver.find_elements_by_xpath('//tbody/tr/td[9]')
# Daerah Penangkapan
daerah_penangkapan = driver.find_elements_by_xpath('//tbody/tr/td[10]')
# Pelabuhan Pangkalan
plbh_pangkln = driver.find_elements_by_xpath('//tbody/tr/td[11]')

```

```python
scarpper_sipiaktif = []
```

```python
for i in range(len(badan_hukum)):
    sipi_dict={'Perorangan / Badan Hukum':badan_hukum[i].text,
              'NOMOR SIUP':nomor_siup[i].text,
              'NAMA KAPAL':nama_kapal[i].text,
              'NOMOR SIPI/SIKPI':nmr_sipi_sikpi[i].text,
              'TANGGAL TERBIT':tanggal_terbit[i].text,
              'TANGGAL BERLAKU':tanggal_berlaku[i].text,
              'ALAT TANGKAP':alat_tangkap[i].text,
              'GT KAPAL':gt_kapal[i].text,
              'TANDA SELAR':tanda_selar[i].text,
              'DAERAH PENANGKAPAN':daerah_penangkapan[i].text,
              'PELABUHAN PANGKALAN':plbh_pangkln[i].text}
    
    scarpper_sipiaktif.append(sipi_dict)
      
    
    
```

```python
# DataFrame
dataFrame_sipi = pd.DataFrame(scarpper_sipiaktif)
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
def check_exists_by_xpath(driver, xpath):
    try:
        driver.find_element(By.XPATH, xpath)
    except NoSuchElementException:
        return False
    return True
```
